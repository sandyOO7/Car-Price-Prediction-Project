```python
# first click on the load more button multiple times 
for i in range(12):
    load=driver.find_element_by_xpath("//button[@class='rui-39-wj rui-3evoE rui-1JPTg']")
    load.click()
    time.sleep(6)
```

```python
#connecting to the chrome driver
driver=webdriver.Chrome(r"C://Chromedriver.exe")
driver.maximize_window()
```

```python
#url to fetch cars belonging to UP
driver.get("https://www.olx.in/uttar-pradesh_g2001176/cars_c84")
time.sleep(5)
```

```python
# clicking load more button 
for i in range(12):
    load=driver.find_element_by_xpath("//button[@class='rui-39-wj rui-3evoE rui-1JPTg']")
    load.click()
    time.sleep(6)
```

```python
#fetch urls of all the products first
url_list1=[]
for i in driver.find_elements_by_xpath("//div[@class='rui-jAIPe']/div/div[5]/div[2]/div/div[2]/ul/li/a"):
    url_list1.append(i.get_attribute("href")) 
```

```python
#check the urls
url_list1[0:2]
```

```python
#number of urls fetched
print(len(url_list1))
```

```python
#check if all the urls are fetched and no data is repeated
import numpy as np
x=np.array(url_list1)
print(len(np.unique(x)))
```

```python
#drop the duplicate urls
url_list1=np.unique(url_list1)
len(url_list1)
```

```python
#create empty lists
brand_list1=[]
model_list1=[]
variant_list1=[]
year_list1=[]
fuel_list1=[]
trans_list1=[]
kms_list1=[]
owners_list1=[]
price_list1=[]
location_list1=[]
```

```python
#scrap the data
for i in url_list1:
    driver.get(i)
    time.sleep(0.10)
    
    try:
        brand_list1.append(driver.find_element_by_xpath(" //span[@data-aut-id='value_make']").text)
    except NoSuchElementException:
        brand_list1.append('-')    
        
    try:   
        model_list1.append(driver.find_element_by_xpath("//div[@class='_3JPEe']/div[2]/div/span[2]").text)
    except NoSuchElementException: 
        model_list1.append('-')
    
    try:
        variant_list1.append(driver.find_element_by_xpath("//div[@class='_3JPEe']/div[3]/div/span[2]").text)
    except NoSuchElementException: 
        variant_list1.append('-')
       
    
    try:
        year_list1.append(driver.find_element_by_xpath("//div[@class='_3JPEe']/div[4]/div/span[2]").text)
    except NoSuchElementException: 
        year_list1.append('-')
        
        
    try:
        fuel_list1.append(driver.find_element_by_xpath("//div[@class='_3JPEe']/div[5]/div/span[2]").text)
    except NoSuchElementException: 
        fuel_list1.append('-')
     
    
    try:
        trans_list1.append(driver.find_element_by_xpath("//div[@class='_3JPEe']/div[6]/div/span[2]").text)
    except NoSuchElementException: 
        trans_list1.append('-')
       
    
    try:
        kms_list1.append(driver.find_element_by_xpath("//div[@class='_3JPEe']/div[7]/div/span[2]").text)
    except NoSuchElementException: 
        kms_list1.append('-')
        
        
    try:   
        owners_list1.append(driver.find_element_by_xpath("//div[@class='_3JPEe']/div[8]/div/span[2]").text)
    except NoSuchElementException: 
        owners_list1.append('-')
        
    try:   
        price_list1.append(driver.find_element_by_xpath("//div[@class='rui-2CYS9']/section/span[1]").text)
    except NoSuchElementException: 
        price_list1.append('-')
       
    
    
    try:
        location_list1.append(driver.find_element_by_xpath("//div[@class='_2A3Wa']/span").text.split(",")[-1])        
    except NoSuchElementException: 
        location_list1.append('-')
       
```

```python
#overview of collected data
print(brand_list1[0:5])
print("\n")
print(model_list1[0:5])
print("\n")
print(variant_list1[0:5])
print("\n")
print(year_list1[0:5])
print("\n")
print(fuel_list1[0:5])
print("\n")
print(trans_list1[0:5])
print("\n")
print(kms_list1[0:5])
print("\n")
print(owners_list1[0:5])
print("\n")
print(price_list1[0:5])
print("\n")
print(location_list1[0:5])
```

```python
#checking length of all the fetched details
print(len(brand_list1))
print(len(model_list1))
print(len(variant_list1))
print(len(year_list1))
print(len(fuel_list1))
print(len(trans_list1))
print(len(kms_list1))
print(len(owners_list1))
print(len(price_list1))
print(len(location_list1))
```

```python
#creating dataframe
df6=pd.DataFrame({
    'Brand':brand_list1,
    'Model':model_list1,
    'Variant':variant_list1,
    'Year':year_list1,
    'Kms driven':kms_list1,
    'Fuel':fuel_list1,
    'Transmission':trans_list1,
    'Number of owners':owners_list1,
    'Location':location_list1,
    'Price':price_list1})
df6
```

```python
#save the dataframe as csv for future use
df6.to_csv("Cars from Uttar Pradesh.csv")
```
